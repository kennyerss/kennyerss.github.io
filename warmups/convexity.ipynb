```python
# Part 3 

%matplotlib inline
import math
from matplotlib import pyplot as plt 
import numpy as np

# Define domain of our graph
x_ = np.linspace(-10, 10)
fig, ax = plt.subplots(1, 1) 
fig, bx = plt.subplots(1,1)

# Make functions to graph 
def f_(z): return -1*np.log(z)
def g_(z): return -1*np.log(1-z)

# Set axes labels
ax.set(xlabel = "z", 
       ylabel = "f(z)")
bx.set(xlabel = "z",
       ylabel = "g(z)")

# Plot functions
ax.plot(x_, f_(x_))
bx.plot(x_, g_(x_))
```
